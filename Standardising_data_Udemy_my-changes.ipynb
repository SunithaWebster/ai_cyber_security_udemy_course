```python
# We start by importing the required libraries and gathering a dataset
import pandas as pd
data = pd.read_csv("file_pe_headers.csv", sep =",")
x = data.drop(["Name", "Malware"], axis=1).to_numpy()
```

```python
# Next standarize x using StandardScaler instance
from sklearn.preprocessing import StandardScaler
x_standardised = StandardScaler().fit_transform(x)
```

```python
print(x_standardised)
```

It is important to standardise features because performance is highly sensitive to the relative scale of features.
